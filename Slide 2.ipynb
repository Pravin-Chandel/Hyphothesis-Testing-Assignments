```python
# we use Anova Test because,it is used to find out whether there exists a statistically significant difference between the ** mean values of more than one group**
```

```python
# Anova ftest statistics: stats.f_oneway(column-1,column-2,column-3,column-4)
P_value=stats.f_oneway(Data.iloc[: ,0],Data.iloc[: ,1],Data.iloc[: ,2],Data.iloc[: ,3])
P_value
```
