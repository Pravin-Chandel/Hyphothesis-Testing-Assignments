```python
obs=np.array([[271,267,269,280],[29,33,31,20]])
obs
```

### Lets,Assume that, 
## H0: Proportion of defective % across the center is same.(if,p_value> 0.05 then will accept H0)
## H1: Proportion of defective % across the center is different.

```python
## Performimg chi2 contengency independence test
```

```python
chi2_contingency(obs)
```

```python
# Here, P-value=0.27,DOF= 3
```
